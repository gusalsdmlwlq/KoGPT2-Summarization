```python
from kogpt2.pytorch_kogpt2 import get_pytorch_kogpt2_model
from transformers import GPT2LMHeadModel, GPT2Config
import torch
```

```python
model, _ = get_pytorch_kogpt2_model()
```

```python
model
```

```python
config = GPT2Config(vocab_size=50000)
```

```python
config
```

```python
ckpt = model.state_dict()
torch.save(ckpt, "save/kogpt2.pt")
```

```python
gpt2 = GPT2LMHeadModel(config)
```

```python
ckpt = torch.load("save/kogpt2.pt", map_location = lambda storage, loc: storage)
gpt2.load_state_dict(ckpt)
```
